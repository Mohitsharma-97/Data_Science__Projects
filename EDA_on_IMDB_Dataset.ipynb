#Problem 4: Make a table of all the producers and directors of each movie. Find the top 3 producers who have produced movies with the highest average RoI?

```python
#creating a df with no na values in crew
imdb_df4=imdb_df2[imdb_df['crew'].notna()]  ##choosing imdf_df2 because this already had roi data
```

```python
len(imdb_df4)
```

```python
# creating a function which returns the list of producers and directors from each movie
def get_producers_and_directors(crew_list):
  try:
    producer_list=[]
    director_list=[]
    for elem in crew_list:

      if elem['job']=='Producer':
        producer_list.append(elem['name'])

      if elem['job']=='Director':
        director_list.append(elem['name'])

    return producer_list,director_list

  except:
    return [],[]
```

```python
get_producers_and_directors(imdb_df4.loc[1,'crew'])
```

```python
imdb_df4[['producers','directors']]=imdb_df4.apply(lambda row: get_producers_and_directors(row['crew']),axis=1,result_type='expand')
```

```python
imdb_df4.head()
```

```python
# creating a dataframe which have movie producer and roi columnns
cols=['Movie','Producer','ROI']
movie_prod_roi=pd.DataFrame(columns=cols)
for index,row in imdb_df4.iterrows():
  for producer in row['producers']:
    dct={'Movie':row['original_title'],'Producer':producer, 'ROI': row['roi']}
    movie_prod_roi=movie_prod_roi.append(dct,ignore_index=True)
```

```python
movie_prod_roi
```

```python
#creating a distinct producer list
distinct_producer_list=list(set(movie_prod_roi['Producer']))
distinct_producer_list
```

```python
#creating a dataframe in which we have producer columns and movie list of producer and list of roi corresponding to movie names present in movie list
cols=['Movie','Producer','Avg_ROI']
movie_prod_avgroi = []
for producer in distinct_producer_list:
  movie_list=[]
  roi_list=[]
  for index,row in movie_prod_roi.iterrows():
    if producer == row['Producer']:
      movie_list.append(row['Movie'])
      roi_list.append(row['ROI'])
  movie_prod_avgroi.append({'Movie':movie_list,'Producer':producer,'ROI':roi_list})
```

```python
movie_prod_avgroi_df=pd.DataFrame(movie_prod_avgroi)
movie_prod_avgroi_df
```

```python
movie_prod_avgroi_df['Avg_ROI']=movie_prod_avgroi_df.apply(lambda row: sum(row['ROI'])/len(row['ROI']),axis=1)
```

```python
movie_prod_avgroi_df.sort_values(['Avg_ROI'],ascending=False,ignore_index=True)
```

```python
top3_prodcuers_with_highest_avg_roi=movie_prod_avgroi_df.sort_values(['Avg_ROI'],ascending=False,ignore_index=True).iloc[0:3]
```

```python
##The top 3 producers who have produced movies with the highest average RoI
top3_prodcuers_with_highest_avg_roi
```

#Problem 5 : Which actor has acted in the most number of movies? Deep dive into the movies, genres and profits corresponding to this actor.

```python
imdb_df5=imdb_df1[imdb_df['cast'].notna()] 
```

```python
# creating a function which returns the list of actors from each movie
def get_actors(cast_list):
  try:
    actor_list=[]
    for elem in cast_list:
      actor_list.append(elem['name'])

    return actor_list
  except:
    return []
```

```python
imdb_df5['actors']=imdb_df5.apply(lambda row: get_actors(row['cast']),axis=1)
```

```python
# creating a dataframe which have movie actor ,genre and proft columnns
cols=['Movie','Actor','Genres','Profit']
df=pd.DataFrame(columns=cols)
for index,row in imdb_df5.iterrows():
  for actor in row['actors']:
    dct={'Movie':row['original_title'],'Actor':actor, 'Genres': row['genres'],'Profit':row['profit']}
    df=df.append(dct,ignore_index=True)
```

```python
df
```

```python
#grouping actors with movie counts
df.groupby('Actor')['Movie'].count().reset_index()
```

```python
#sorting the above data frame by descending values
df.groupby('Actor')['Movie'].count().reset_index().sort_values(['Movie'],ignore_index=True,ascending=False)
```

```python
#actor with highest movies
actor_highest_movies=df.groupby('Actor')['Movie'].count().reset_index().sort_values(['Movie'],ignore_index=True,ascending=False).loc[0:1,'Actor'].tolist()
actor_highest_movies
```

## There are two actors with hoghest number of movies and they are Samuel L. Jackson and Robert De Niro. We derive the required info of both actors separately 

```python
# Required info about Samuel L. jackson
cols=['Movie','Genres','Profit']
Samuel_L_Jackson_info=pd.DataFrame()
for index,rows in df.iterrows():
  if rows['Actor']==actor_highest_movies[0]:
    Samuel_L_Jackson_info=Samuel_L_Jackson_info.append({'Movie':rows['Movie'],'Genres':rows['Genres'],'Profit':rows['Profit']},ignore_index=True)
Samuel_L_Jackson_info
```

```python
# Required info about Robert De Niro
cols=['Movie','Genres','Profit']
Robert_De_Niro_info=pd.DataFrame()
for index,rows in df.iterrows():
  if rows['Actor']==actor_highest_movies[0]:
    Robert_De_Niro_info=Robert_De_Niro_info.append({'Movie':rows['Movie'],'Genres':rows['Genres'],'Profit':rows['Profit']},ignore_index=True)
Robert_De_Niro_info
```
